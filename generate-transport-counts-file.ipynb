```python
import pandas as pd
```

```python
sp = pd.read_csv("/Users/mike/witan.send/data/version-A/send-population.csv")

sp['state'] = sp['need'] + '-' + sp['setting']
sp.head()
```

```python
df = sp.groupby('state').sum()
df = df.drop(['calendar-year', 'academic-year'], axis=1)
df.head(10)
```

```python
internal = pd.read_csv("/Users/mike/witan.send/data/version-A/raw/Transport-2017-Internal-Version-A-.csv")
internal['state'] = internal['Need'] + '-' + internal['Setting']
df1 = internal.groupby('state').sum().drop(['A/Y'], axis=1)
df1.head(10)
```

```python
ex = pd.read_csv("/Users/mike/witan.send/data/version-A/raw/Transport-2017-External-Version-A-.csv")
ex['state'] = ex['Need'] + '-' + ex['Setting']
df2 = ex.groupby('state').sum().drop(['A/Y'], axis=1)
df2.head(10)
```

```python
df = df.rename(columns={"population": "none"})
df['internal'] = df1['Headcount']
df['external'] = df2['Headcount']
df = df.fillna(0)
```

```python
df['none'] = df['none'] - df['internal'] - df['external']
```

```python
df['none-proportion'] = df.none / (df.none + df.internal + df.external)
df['internal-proportion'] = df.internal / (df.none + df.internal + df.external)
df['external-proportion'] = df.external / (df.none + df.internal + df.external)
```
